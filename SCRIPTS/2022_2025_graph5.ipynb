```python
# This script creates a graph specifically off the years 2022-2025
# Make sure you have the preprocessed data (preprocssed_data.csv) uploaded to this file
```

```python
import pandas as pd

df = pd.read_csv('/content/preprocessed_data.csv')
display(df.head())
```

```python
# Load the file `preprocessed_data.csv` into a dataframe and create a line plot for the United States for the years 2022-2025.
df_filtered = df[df['State'] == 'United States'][['State', '2022', '2023', '2024', '2025']]
display(df_filtered)
```

```python
# Reshape the filtered data from wide to long format to prepare it for plotting.
df_long = df_filtered.melt(id_vars=['State'], var_name='Year', value_name='Price of Electricity (cents per kilowatthour)')
display(df_long)
```

```python
# Create a line plot using the melted data frame to visualize the trend for the United States from 2022-2025 with the appropriate labels and titles

import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(df_long['Year'], df_long['Price of Electricity (cents per kilowatthour)'])
plt.title('United States Average Price of Electricity Per Year (2022-2025)')
plt.xlabel('Year')
plt.ylabel('Price of Electricity (cents per kilowatthour)')
plt.show()
```
