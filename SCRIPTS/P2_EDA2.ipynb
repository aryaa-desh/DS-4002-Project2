```python
# This script performs preliminary EDA
# Access to the preprocessed data csv (preprocessed_data.csv) is required for running this script
```

```python
import pandas as pd

df = pd.read_csv('/content/preprocessed_data.csv')
display(df.head())
```

```python
# Filter the df to select the row where the "State" column is "United States"
df_us = df[df['State'] == 'United States']
display(df_us)
```

```python

# reshape the data so that years are in one column and electricity costs in another
df_us_melted = df_us.melt(id_vars='State', var_name='Year', value_name='Electricity Cost')
df_us_melted['Year'] = pd.to_numeric(df_us_melted['Year'])
display(df_us_melted.head())
```

```python
# Creating the graph
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker

df_us_melted_filtered = df_us_melted[df_us_melted['Year'] <= 2021]

plt.figure(figsize=(12, 6))
plt.plot(df_us_melted_filtered['Year'], df_us_melted_filtered['Electricity Cost'])
plt.title("United States Average Electricity Cost Over Time")
plt.xlabel("Year")
plt.ylabel("Average Electricity Cost (cents/kWh)")
plt.grid(True)

# Set x-axis ticks to be whole numbers
plt.gca().xaxis.set_major_locator(mticker.MaxNLocator(integer=True))

plt.show()
```

```python
# Extract Texas data
df_texas = df[df['State'] == 'Texas']
display(df_texas)
```

```python
# Reshape the data so that years are in one column and electricity costs in another.
df_texas_melted = df_texas.melt(id_vars='State', var_name='Year', value_name='Electricity Cost')

# Convert the 'Year' column to numeric
df_texas_melted['Year'] = pd.to_numeric(df_texas_melted['Year'])

# Display the head of the melted DataFrame
display(df_texas_melted.head())
```

```python
# Create the graph
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker

df_texas_melted_filtered = df_texas_melted[df_texas_melted['Year'] <= 2021]

plt.figure(figsize=(12, 6))
plt.plot(df_texas_melted_filtered['Year'], df_texas_melted_filtered['Electricity Cost'])
plt.title("Texas Average Electricity Cost Over Time (Up to 2021)")
plt.xlabel("Year")
plt.ylabel("Average Electricity Cost (cents/kWh)")
plt.grid(True)

# Set x-axis ticks to be whole numbers
plt.gca().xaxis.set_major_locator(mticker.MaxNLocator(integer=True))

plt.show()
```
